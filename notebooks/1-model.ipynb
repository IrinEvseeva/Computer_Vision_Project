```python
import cv2
from ultralytics import solutions

# Load the image
image_path = "e3b62e2a0851e988f534736b38f0165e.png"
image = cv2.imread(image_path)
assert image is not None, "Error reading image file"

# Init ObjectBlurrer with your 'yolo12' model
blurrer = solutions.ObjectBlurrer(
    show=True,               # Display the output
    model="last.pt",        # Your face detection model
    blur_ratio=0.15,
    line_width=1,
    classes=[0],              # Assuming class '0' corresponds to faces in your model
)

# Process the image to detect and blur faces
results = blurrer(image)

# Save the output image
output_path = "face_blurring_output.jpg"
cv2.imwrite(output_path, results.plot_im)
```

```python
# Process results list
for result in results:
    boxes = result.boxes  # bbox нужен для детекции
    masks = result.masks  # Маска нужна для сегментации
    keypoints = result.keypoints  # Для определения позы
    probs = result.probs  # Для классификации найденного объекта
    obb = result.obb  # Ориентированный bbox
    result.show()  # display to screen
```
